Overview:
Simple algo using SMA's
The idea is that when the 50d SMA is below the 200d SMA the algo is short on the asset, and when above the algo is long.
Algo is always holding a position in the market and returns are compounting - using full leverage at any given time. No risk management.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

Download three years of apple data from api

```python
aapl = yf.download('AAPL', start='2020-01-01', end='2023-01-01')
```

Calculate SMA's

```python
aapl['SMA_50']=aapl['Close'].rolling(window=50).mean()
aapl['SMA_200']=aapl['Close'].rolling(window=200).mean()
```

Create buy and sell signals for SMA crossover strategy
- buy signal generated when short term SMA crosses above long term SMA, and sell when below
(starting from 201st day as its the first day on which both SMA's have sufficient data)

```python
aapl['Signal']=0
aapl['Signal'][200:]=np.where(aapl['SMA_50'][200:]>aapl['SMA_200'][200:],1,-1)
```

Backtest: calculate the daily returns
- shift in data to align trading signal with day of execution
      e.g. signal generated at close D1 would be executed at close D2

```python
aapl['Market_Returns']=aapl['Close'].pct_change()
aapl['Strategy_Returns']=aapl['Market_Returns']*aapl['Signal'].shift(1)
```

Plot market vs strategy returns with shading to show long/short

```python
plt.figure(figsize=(10,5))
plt.plot(aapl.index[200:], (1 + aapl['Market_Returns'][200:]).cumprod(), label='Market Returns', color='orange')
plt.plot(aapl.index[200:], (1 + aapl['Strategy_Returns'][200:]).cumprod(), label='Strategy Returns', color='blue')
long_positions = aapl['Signal'][200:] > 0
short_positions = aapl['Signal'][200:] < 0
plt.fill_between(aapl.index[200:], 1, (1 + aapl['Strategy_Returns'][200:]).cumprod(), where=long_positions, color='green', alpha=0.3)
plt.fill_between(aapl.index[200:], 1, (1 + aapl['Strategy_Returns'][200:]).cumprod(), where=short_positions, color='red', alpha=0.3)
plt.legend()
plt.show()
```

Comments:Strategy buys into a rising market initially, then takes on almost exclusively short position, underperforming for a period, but producing same return approxiately over the time period

Summary stats

```python
total_return = (1 + aapl['Strategy_Returns'][200:]).cumprod().iloc[-1] - 1
trading_days = 252
annualized_return = (1 + total_return)**(trading_days / len(aapl-200)) - 1
cumulative_returns = (1 + aapl['Strategy_Returns'][200:]).cumprod()
max_drawdown = (cumulative_returns.cummax() - cumulative_returns).max()
sharpe_ratio = aapl['Strategy_Returns'].mean() / aapl['Strategy_Returns'].std() * np.sqrt(trading_days)
print(f"Total Return: {total_return:.2%}")
print(f"Annualized Return: {annualized_return:.2%}")
print(f"Maximum Drawdown: {max_drawdown:.2%}")
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
```

Brief reflections:
Unsuprisingly not a particularly successful strategy compared to the returns of the stock
Doesn't account for transaction costs, slippage, market impact
Additionally, could add more indicators, risk management, or test across more stocks and timeframes
